Tesnsors 
- A tensor is a mathematical object that generalizes scalars, vectors, and matrices to higher dimensions, acting as a container for data in n-dimensions

```python
create_tensor = torch.tensor([1, 2, 3])
print(create_tensor)
create_tensor.dtype
```

```python
create_tensor = torch.tensor([1, 2, 3],dtype=torch.float32)
print(create_tensor)
create_tensor.dtype
```

```python
#random tensor
torch.rand(3, 4)
```

```python
# stop update random tensor
torch.manual_seed(42)
torch.rand(3, 4)
```

```python
torch.manual_seed(42)
torch.rand(3, 4)
```

Both Random tensors are same no furter updates on manual_seed 42

```python
torch.eye(5)
```

```python
torch.ones(3,4)
```

```python
torch.add(torch.ones(3,3),torch.eye(3))
```

```python
torch.mul(torch.ones(3,3),torch.eye(3)) # element wise multiplication
```

```python
torch.matmul(torch.ones(3,3),torch.eye(3,3)) # matrix multiplication
```

```python
torch.matmul(torch.ones(2,3),torch.eye(3,3)) # matrix multiplication with non square matrix
```

```python
x = torch.ones(3,4)
y = torch.zeros_like(x) # dimensions like x 
```

```python
print('x = ',x)
print('y = ',y)
```

```python
x.to(dtype= torch.int32)
```

```python
torch.mean(torch.tensor([1,2,3,4],dtype=torch.float32))
```

```python
torch.median(torch.tensor([1,2,3,4],dtype=torch.float32))
```

```python
torch.mode(torch.tensor([1,2,4,4],dtype=torch.float32))
```

```python
torch.argmax(torch.tensor([1,2,4,4],dtype=torch.float32)) # index of max value
```

```python
torch.argmin(torch.tensor([1,2,4,4],dtype=torch.float32)) # index of min value
```
